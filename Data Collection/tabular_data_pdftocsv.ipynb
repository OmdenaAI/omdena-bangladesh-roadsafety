```python
# import libraries
import tabula
import pandas as pd
import numpy as np
```

### [Road safety Facts 2014](https://ari.buet.ac.bd/wp-content/uploads/2020/09/Road-Safety-Facts-2014.pdf)

```python
file = "https://ari.buet.ac.bd/wp-content/uploads/2020/09/Road-Safety-Facts-2014.pdf"
tables = tabula.read_pdf(file, pages = "all", multiple_tables = True)
```

```python
# Definig a function to return each table as a DataFrame and take its first row as the header

def table_with_header(input_df):
    """
    This function returns DataFrame for 
    selected table
    """
    
    df = input_df
    df.columns = df.iloc[0]
    df = df[1:]

    return df
```

#### TABLE 1: NUMBER OF ACCIDENT BY ACCIDENT SEVERITY

```python
df_0 = table_with_header(tables[0])
df_0
```

#### TABLE 2: NUMBER OF CASUALTIES BY CASUALTY INJURY

```python
df_1 = table_with_header(tables[1])
df_1
```

#### TABLE 3: NUMBER OF ACCIDENTS BY COLLISION TYPE

```python
df_2 = table_with_header(tables[2])
df_2.columns = ['Year', 'Head On', 'Rear End', 'Right Angle', 'Side Swipe', 'Overturn', 'Hit Object in Road', \
               'Hit Object off Road', 'Hit Parked Vehicle', 'Hit Pedestrian', 'Hit Animal', 'Other', 'TOTAL' ]
df_2
```

#### TABLE 4: NUMBER OF CASUALTIES BY COLLISION TYPE

```python
df_3 = table_with_header(tables[3])
df_3.columns = ['Year', 'Head On', 'Rear End', 'Right Angle', 'Side Swipe', 'Overturn', 'Hit Object in Road', \
               'Hit Object off Road', 'Hit Parked Vehicle', 'Hit Pedestrian', 'Hit Animal', 'Other', 'TOTAL' ]
df_3
```

#### TABLE 5: NUMBER OF DRIVER, PASSENGER AND PEDESTRIAN CASUALTIES BY YEAR

```python
df_4 = table_with_header(tables[4])
df_4.columns = ['Year', 'Fatal_driver', 'Grievous_driver', 'Simple_driver', 'TOTAL_driver', 'Fatal_passenger', 'Grievous_passenger',
       'Simple_passenger', 'TOTAL_passenger', 'Fatal_pedestrian', 'Grievous_pedestrian', 'Simple_pedestrian', 'TOTAL_pedestrian']
df_4
```
